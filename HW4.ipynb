```python
import re
hashtags= []
for t in tweets:
    txt = t.text
    arr = re.findall(r"#(\w+)", txt)
    hashtags.extend(arr)
```

```python
htdic = {}
for ht in hashtags:
    try:
        htdic[ht] += 1
    except:
        htdic[ht] = 1
```
