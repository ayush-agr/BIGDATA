```python
cntyr = {}
for index,row in tble.iterrows():
    yr = int(row['iyear'])
    if(yr>=1970 and yr<=2017):
        try:
            cntyr[yr]+=1
        except:
            cntyr[yr]=1
            
print(cntyr)
```

```python
plt.plot(cntyr.keys(),cntyr.values())
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.title('Distribution of frequency with crime year')
plt.rcParams['figure.figsize'] = [30, 5]
plt.show()
```

```python
mx = 0
mxyr = ""
for i in cntyr:
    if(cntyr[i] >= mx):
        mx = cntyr[i]
        mxyr = i
print(mxyr)
```

### Hence year with maximum crime is 2014.

```python
freqNA = {}
for index,row in tble.iterrows():
    rg = row['region_txt']
    yr = int(row['iyear'])
    if(rg == "North America"):
        try:
            freqNA[yr]+=1
        except:
            freqNA[yr]=1
```

```python
plt.plot(freqNA.keys(),freqNA.values())
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.title('Distribution of frequency with crime in north america with year')
plt.rcParams['figure.figsize'] = [30, 5]
plt.show()
```

```python
import operator
mx = max(freqNA.items(),key=operator.itemgetter(1))[0]
print(mx)
freqNA[mx]
```

```python
yrs = 1970
yre = 1979
frq = {}
for i in freqNA:
    if(i>=yrs and i<=yre):
        try:
            frq[yrs]+=1
        except:
            frq[yrs]=1
    else:
        yrs+=10
        yre+=10
```

```python
frq
```

```python
plt.plot(frq.keys(),frq.values())
plt.xlabel('Decade')
plt.ylabel('Frequency')
plt.title('Distribution of frequency with crime decade')
plt.rcParams['figure.figsize'] = [30, 5]
plt.show()
```

### Q2.)In 2014 there was the most number of terrorist attacks.

```python
freqcnt = {}
for index,row in tble.iterrows():
    yr = int(row['iyear'])
    cnt = row['country_txt']
    if(yr==2014):
        try:
            freqcnt[cnt]+=1
        except:
            freqcnt[cnt]=1
```

## Q2.) Iraq has the most number of terrorist attacks in 2014

```python
summaries = []
motives = []
for index,row in tble.iterrows():
    smr = row['summary']
    mtv = row['motive']
    summaries.append(smr)
    motives.append(mtv)
```

## Word cloud of summaries

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
import math
import numpy as np
sep = ":"
newsumm = []
for s in summaries:
    if(type(s) != float):
        rest = s.split(sep,1)[-1]
        newsumm.append(rest)
```

```python
newsumm
```

```python
words = []
for sm in newsumm:
    tokens = sm.split()
    for i in range(len(tokens)):
        tokens[i]= tokens[i].lower()
    words.extend(tokens)
```
